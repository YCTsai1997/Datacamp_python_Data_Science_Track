```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

## Hue
 - subgroups with color (hue)
 - You can set the "hue" parameter equal to the DataFrame column name "smoker" and then Seaborn will automatically color each point by whether they are a smoker. Plus, it will add a legend to the plot automatically! If you don't want to use Pandas, you can set it equal to a list of values instead of a column name.

```python
## A basic scatter plot
sns.scatterplot(
    x = "total_bill", ## in seaborn you don't have to indicate the dataset to extract the column
    y = "tip",
    data = tips, 
    hue = "smoker"
)
plt.show()
```

```python
## Setting hue order
sns.scatterplot(
    x = "total_bill", ## in seaborn you don't have to indicate the dataset to extract the column
    y = "tip",
    data = tips, 
    hue = "smoker",
    hue_order = ["No", "Yes"]
)
plt.show()
```

```python
## Setting hue order
hue_colors = {
    "Yes": "black",
    "No": "red"
}

sns.scatterplot(
    x = "total_bill", 
    y = "tip",
    data = tips, 
    hue = "smoker",
    palette = hue_colors ## change the color use palette! Takes dictionary only!
)
plt.show()
```

```python
## Using HTML hex color codes with hue 

hue_colors = {
    "Yes": "#808080",
    "No": "#00FF00"
}

sns.scatterplot(
    x = "total_bill", 
    y = "tip",
    data = tips, 
    hue = "smoker",
    palette = hue_colors ## change the color use palette!
)
plt.show()
```

```python
## Using hue with count plots 
sns.countplot(
    x="smoker",
    data=tips,
    hue="sex"
)
plt.show()
```

## Introduction to relational plots and subplots
### Introducing replot()
- replot() stands for "relational plot" and enables you to visualize the relationship between two quantitative variables using either scatter plots or line plots. 
- why use relplot() instea of scatterplot()?
    - relplot() lets you create subplots in a single figure 
    

```python
sns.relplot(
    x = "total_bill", 
    y = "tip",
    data = tips, 
    kind = "scatter"
)
plt.show()
```

```python
sns.relplot(
    x = "total_bill", 
    y = "tip",
    data = tips, 
    kind = "scatter",
    col = "smoker"
)
plt.show()
```

```python
sns.relplot(
    x = "total_bill", 
    y = "tip",
    data = tips, 
    kind = "scatter",
    row = "smoker"
)
plt.show()
```

```python
sns.relplot(
    x = "total_bill", 
    y = "tip",
    data = tips, 
    kind = "scatter",
    col = "smoker",
    row="time"
)
plt.show()
```

### Wrapping columns

```python
sns.relplot(
    x = "total_bill", 
    y = "tip",
    data = tips, 
    kind = "scatter",
    col = "day",
    col_wrap= 4
)
plt.show()
```

### Ordeing columns

```python
sns.relplot(
    x = "total_bill", 
    y = "tip",
    data = tips, 
    kind = "scatter",
    col = "day",
    col_wrap= 2,
    col_order =["Thur","Fri","Sat","Sun"]
)
plt.show()
```

- whis =2.0: meaning change the default IQR from 1.5 to 2 times the IQR
- whis = [5,95]: show the 5th and 95th percentiles
- whis = [0, 100]: show min and max value

## Point plots
- points show mean of quantitative variable
- vertical lines show 95% confidence intervals
- Difference between point plots vs. line plots: 
    - Line plot has quantitative variable on x axis
    - Point plot has categorical variable on x-axis

```python
sns.catplot(
    x = "smoker", 
    y = "total_bill",
    data = tips, 
    kind = "point"
)
plt.show()
```

```python
sns.catplot(
    x = "smoker", 
    y = "total_bill",
    data = tips, 
    kind = "point",
    estimator = np.median # To have the points confidence intervals be calculated for the median instead of the mean
) # Median is more robust to outliers, so if your dataset has a lot of outliers, 
        # the median may be a better statistic to use. 
plt.show()
```

## Customizing the confidence intervals 

```python
sns.catplot(
    x = "smoker", 
    y = "total_bill",
    data = tips, 
    kind = "point",
    capsize = 0.2
) 
plt.show()
```

## Palettes
 - Seaborn uses the set_palette() function to define a palette

```python
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    sns.distplot(tui['net_cost'])
plt.show()
```

## Displaying palettes
 - sns.palplot() function displays a palette
 - sns.color_paletter() returns the current palette

```python
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    sns.palplot(sns.color_palette())
plt.show()
```
